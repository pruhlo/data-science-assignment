### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

def describe_range(lst):
    import numpy as np
    lst = np.array(lst)
    minimum = np.min(lst)
    maximum = np.max(lst)
    avg = np.mean(lst)
    med = np.median(lst)
    std = np.std(lst)
    return minimum, maximum, avg, med, std
print(describe_range(list_with_random_numbers))
```

```python
list_with_random_numbers.count(1)
unique_digitals = set(list_with_random_numbers)
d_count = {}
for i in unique_digitals:
    d_count[str(i)] = list_with_random_numbers.count(i)
print(d_count)
```
