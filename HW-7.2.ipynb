## 1 Менше або більше середнього
Створіть список із $20$ випадкових цілих чисел із відрізка $[-100, 100]$.  
Виведіть такі дані:
- середнє арифметичне від усіх значень
- значення, що менші середнього
- значення, що дорівнюють середньому (якщо такі є)
- значення, що більші середнього

Перед кожним набором значень виведіть відповідний рядок (коментар, опис наступних значень).

**Зауваження.** Усі списки формувати за допомогою **_list comprehension_**.

```python
# Місце для Вашого коду
import random
import numpy as np
liste = [random.randint(-100,100) for i in range(20)]
mean = np.mean(liste)
smaller_mean = [i for i in liste if i<mean]
if_mean = [i for i in liste if i==mean]
greater_mean = [i for i in liste if i>mean]
# print(f'{liste=}') не працює, так як python3.7 , повинен бути 3.8 і вище
print(f'liste = {liste}')
print(f'mean = {mean}')
print(f'smaller_mean = {smaller_mean}')
print(f'if_mean = {if_mean}')
print(f'greater_mean = {greater_mean}')
```

## 2 Практика **_list/dictionary comprehension_**
Дано такі змінні.

```python
nums = [i for i in range(1,1001)]
string = "Practice Following Problems to Drill List Comprehension in Your Head."
```

Розв'яжіть наступні задачі для цих змінних за допомогою **_list/dictionary comprehension_**.

#### **1)** Знайдіть усі числа від $1$ до $1000$, які діляться на $8$.

```python
# Місце для Вашого коду
multiples8 = [i for i in nums if i%8==0]
print(multiples8)
```

#### **2)** Знайдіть усі числа від $1$ до $1000$, які містять цифру $6$ у їх записі.

```python
# Місце для Вашого коду
has6 = [i for i in nums if str(i).count('6')>=1]
print(has6)
```

#### **3)** Знайдіть усі слова, які складаються із менше ніж $5$ букв, у реченні зі змінної __string__.

**Зауваження.** Зверніть увагу на метод __split__.

```python
# Місце для Вашого коду
word_less_5 = [i for i in string.split() if len(i)<5] # з урахуванням крапки у кінці, ще варіант попередньо убрати пунктуацію
print(word_less_5)
```

#### **4)** Порахуйте довжину кожного слова у реченні зі змінної __string__.

```python
# Місце для Вашого коду
lenth_words = {str(i):len(i) for i in string.split()}
print(lenth_words)
```

#### **5)*** Знайдіть усі числа від $1$ до $1000$, які діляться на будь-яке із чисел $13$, $23$, $37$.

```python
# Місце для Вашого коду
multiples_13_23_37 = [i for i in nums if i%13==0 or i%23==0 or i%37==0]
print(multiples_13_23_37)
```

#### **6)*** Для кожного числа від $1$ до $1000$ знайдіть найбільше одноцифрове число ($1, \ldots, 9$), на яке воно ділиться.

```python
# Місце для Вашого коду
maximum_divisor_1_9 = {x:max([i for i in range(1,10) if x%i==0]) for x in nums}
print(maximum_divisor_1_9)
```
